## Matplotlib
### 데이터들을 그래프로 시각화
### http://matplotlib.org

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## 라인 그래프 그리기

```python
x = [10, 14, 19, 20, 25]
plt.plot(x)
plt.show()
```

```python
x = np.arange(-4.5, 5, 0.5)
y = x ** 2
plt.plot(x, y)
plt.show()
```

## 한번에 여려 그래프 그리기

```python
x = np.arange(-4.5, 5, 0.5)

y1 = x**2
y2 = x+30
y3 = x**2+10

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

```python
# 참고. x, y 길이가 다를 때
list_y1 = list(y1)
list_y1.append(22)
np_y1 = np.array(list_y1)
```

```python
plt.plot(x, np_y1)
plt.show()
```

## plot 옵션

```python
x = np.arange(-4.5, 5, 0.5)

y1 = x**2
y2 = x+30
y3 = x**2+10

plt.plot(x, y1, color="red", linestyle="dashed")
plt.plot(x, y2, "m:")
plt.plot(x, y3, color='#00FF00', linestyle="-.")

plt.show()
```

## 축의범위조정

```python
x = np.linspace(0, 10, 100)
y = x ** 2

plt.plot(x, y, '--r')
plt.xlim(-20, 30)
# plt.ylim(-20, 120)
# plt.axis([-20, 30, -20, 120])
plt.show()
```

## label 과 legend

```python
# matplotlib 폰트설정
# plt.rc('font', family='AppleGothic') # For MacOS
# plt.rc('font', family='Malgun Gothic') # For Windows

x = np.linspace(0, 10, 100)
y1 = x ** 2
y2 = x ** 3

plt.plot(x, y1, 'r', label='line-1')
plt.plot(x, y2, 'g', label='line-2')

plt.xlabel('x-label')
plt.ylabel('y-label')

# plt.legend(loc=8)
plt.legend(bbox_to_anchor=(1.04, 1))
plt.show() 
```

## subplot

```python
plt.subplot(131)
plt.plot([1, 2], color='r')

plt.subplot(132)
plt.plot([1, 2], color='g')

plt.subplot(133)
plt.plot([1, 2], color='b')

plt.show()
```

```python
plt.subplot(221)
plt.plot([1, 2], color='r')

plt.subplot(222)
plt.plot([1, 2], color='g')

plt.subplot(223)
plt.plot([1, 2], color='b')

plt.subplot(224)
plt.plot([1, 2], color='r')

plt.show()
```

## Gird

```python
plt.figure(figsize=(10, 5))

grid = plt.GridSpec(2, 3)

plt.subplot(grid[0, 0])
plt.plot([1, 2], color='r')

plt.subplot(grid[0, 1])
plt.plot([1, 2], color='g')

plt.subplot(grid[0, 2])
plt.plot([1, 2], color='b')

plt.subplot(grid[1, :2])
plt.plot([1, 2], color='b', linestyle="--")

plt.subplot(grid[1, 2])
plt.plot([1, 2], color='r', linestyle=":")

plt.show()
```

## Scatter (산점도)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(30)
y = np.random.rand(30)
a = np.random.rand(30)
b = np.random.rand(30)
e = np.random.rand(30)
f = np.random.rand(30)
color = np.random.rand(30)

size = 100 * np.random.rand(30)

plt.scatter(x, y, s=size, c=color, alpha=0.5, marker='o')
plt.scatter(a, b, s=size, c=color, alpha=0.5, marker='^')
plt.scatter(e, f, s=size, c=color, alpha=0.5, marker='+')
plt.show()
```

## Histogram (히스토그램)

```python
data = [np.random.randint(1, 7) for i in range(10)]
print(data)
```

```python
# data = [np.random.randint(1, 7) for i in range(10)]
plt.hist(data, bins=6, alpha=0.5)
plt.show()
```
